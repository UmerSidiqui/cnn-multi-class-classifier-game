```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Flatten,Conv2D,MaxPooling2D,BatchNormalization,Dropout
```

```python
model = Sequential()

model.add(Conv2D(8,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(256,256,3)))

model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Conv2D(16,kernel_size=(3,3),padding='valid',activation='relu'))

model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu'))

model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Flatten())
model.add(Dense(32,activation='relu'))

model.add(Dense(8,activation='relu'))

model.add(Dense(5,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(
    optimizer="adam",
    loss='categorical_crossentropy',
    metrics=['accuracy']
  )
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Define the EarlyStopping callback
callback = EarlyStopping(
    monitor='accuracy', # or 'val_accuracy' if you want to monitor validation accuracy
    mode='auto', # 'auto' determines whether the monitored quantity should be maximized or minimized automatically
    patience=0, # How many epochs to wait after the condition is met. 0 means the training will stop immediately after reaching the threshold.
    verbose=1, # To print a message when the callback triggers.
    baseline=0.95 # Threshold for stopping, but note: 'threshold' is not a valid argument for EarlyStopping
)
```

```python
history = model.fit(train_images,train_labels,batch_size=75,epochs=7,callbacks=callback)
```
