```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Flatten
from keras.applications.xception import Xception
```

```python
conv_base = Xception(
    weights='imagenet',
    include_top = False,
    input_shape=(256,256,3)
)
```

```python
conv_base.trainable = True

set_trainable = False

for layer in conv_base.layers:
  if layer.name == 'block14_sepconv1':
    set_trainable = True
  if set_trainable:
    layer.trainable = True
  else:
    layer.trainable = False

for layer in conv_base.layers:
  print(layer.name,layer.trainable)
```

```python
conv_base.summary()
```

```python
model = Sequential()

model.add(conv_base)
model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(Dense(5,activation='softmax'))
```

```python
model.compile(
    optimizer=keras.optimizers.RMSprop(learning_rate=1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy']
  )
```

```python
import numpy as np
train_images=np.load("train_images2.npy")
train_labels=np.load("train_labels2.npy")
```

```python
history = model.fit(train_images,train_labels,batch_size=10,epochs=1)
```

```python
model.save("multi class model 7")
```

```python
import cv2
import numpy as np
import mediapipe as mp

mpdrawing=mp.solutions.drawing_utils
mphands=mp.solutions.hands


cpt=cv2.VideoCapture(0)
with mphands.Hands(min_detection_confidence=0.8,min_tracking_confidence=0.5) as hands:
    while True:
        ret,frame=cpt.read()
        img=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        img.flags.writeable=False
        res=hands.process(img)
        img.flags.writeable=True
        img=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
        if res.multi_hand_landmarks:
            for num,hand in enumerate(res.multi_hand_landmarks):
                mpdrawing.draw_landmarks(img,hand,mphands.HAND_CONNECTIONS)

        img= img.astype('float32') / 255.0 
        img=cv2.resize(img,(256,256))
        cv2.imshow("win",img)

         
        img = np.expand_dims(img, axis=0)
        res=model.predict(img)
        a=res.argmax()
        if a==0:
            print("down")
        elif a==1:
            print("fist")
        elif a==2:
            print("left")
        elif a==3:
            print("right")
        elif a==4:
            print("up")


        
        if cv2.waitKey(250)==ord('p'):
            break

    cpt.release()
    cv2.destroyAllWindows()
```
